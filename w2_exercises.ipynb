```python
# Week 2 Quiz: Data Visualization

## Starter Code (RUN THIS FIRST IN YOUR OWN ENVIRONMENT)

import pandas as pd
import matplotlib.pyplot as plt

data = {
    "genre": [
        "Pop","Rock","Hip-Hop","Electronic","Pop","Rock","Hip-Hop","Electronic",
        "Pop","Rock","Hip-Hop","Electronic","Pop","Rock","Hip-Hop","Electronic",
        "Pop","Rock","Hip-Hop","Electronic"
    ],
    "streams": [
        120,90,150,70,130,95,140,80,
        110,85,155,75,125,100,145,78,
        118,92,160,82
    ],
    "likes": [
        30,20,50,15,35,22,48,18,
        28,19,55,16,32,25,47,17,
        29,21,60,19
    ],
    "shares": [
        10,8,20,5,12,9,18,6,
        9,7,22,5,11,10,19,6,
        10,8,25,7
    ],
    "year": [
        2020,2020,2020,2020,2021,2021,2021,2021,
        2022,2022,2022,2022,2023,2023,2023,2023,
        2024,2024,2024,2024
    ]
}

df = pd.DataFrame(data)
```

```python
## 1. Create a Bar Plot

 #x-axis: genre  
 #y-axis: streams  
 #title: Streams by Genre  
plt.figure(figsize=(8,6))
strms_genre = df.groupby('genre')['streams'].mean()
# i wanted to sort the values but the output didnt sort them
# keeping it fundamentally the same
#added some flare though :) (mostly colors and font sizes!)


plt.bar(strms_genre.index, strms_genre.values, color=  "#f57baa", width= 0.4)

#output has all the ax.spines so ill leave that alone 


plt.xlabel("Genre", fontsize= 14, color = "#095029")
plt.ylabel("Streams", fontsize= 14, color = "#095029")
plt.title("Streams by Genre", fontsize= 18, color = "#0B5858")
plt.xticks(rotation=90)


plt.tight_layout()
plt.show()
```

```python

## 2. Create a Scatter Plot
#x-axis: likes
#y-axis: streams
#title: Streams vs Likes

plt.figure(figsize=(10,8))

plt.scatter(
    df['likes'],
    df['streams'],
    color= "#1D8A4E",
    alpha=0.7
    )

plt.xlabel("Likes", fontsize= 14, color = "#5A5102")
plt.ylabel("Streams", fontsize= 14, color = "#0E107C")
plt.title("Streams vs. Likes", fontsize= 18, color = "#54018B")

#same here, no ax.spines and a little color and font size changes

plt.show()
```

```python
## 3. Create a Histogram
#Use streams column
#title: Streams Distribution
plt.figure(figsize=(9,6))

plt.hist(
    df['streams'],
    color= "#89fa9c",
    edgecolor= "#dd2a57")

plt.title("Streams Distribution", fontsize= 22, pad= 20, color = "#531358")


### again: no ax.spines adjusting
#but i did add some colors
plt.show()

```
